```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dfs= pd.read_csv("/content/drive/MyDrive/DATASET /automobile.csv")
```

```python
dfs.shape
```

```python
dfs.head()
```

```python
dfs.replace('?',np.nan,inplace=True)
dfs.head()
```

```python
dfs.shape
```

```python
dfs.dropna(subset=['price'],axis=0,inplace=True)
```

```python
dfs.shape
```

```python
dfs.drop_duplicates(inplace=True)
```

```python
dfs['normalized-losses'].fillna(dfs['normalized-losses'].astype(float).mean(),inplace=True)
```

```python
dfs.dtypes
```

```python
dfs[['normalized-losses','num-of-doors','num-of-cylinders','bore','stroke','horsepower','peak-rpm','price']]=dfs[['normalized-losses','num-of-doors','num-of-cylinders','bore','stroke','horsepower','peak-rpm','price']].astype(float)
```

```python
dfs.dtype()
```

```python
dfs.columns
```

```python
# 1. How does the fuel economy (city-mpg and highway-mpg) vary between different body styles and fuel types?
# Which body style and fuel type have the highest and lowest fuel economy?
# Solution:
plt.subplot(2,1,1)
sns.boxplot(data=dfs,x='body-style',y='city-mpg',hue='fuel-type')
plt.title('city-mpg Vs body styles')

plt.subplot(2,1,2)
sns.boxplot(data=dfs,x='body-style',y='highway-mpg',hue='fuel-type')
plt.title('highway-mpg Vs body styles')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=dfs,x='body-style',y='highway-mpg',hue='fuel-type')
plt.title('highway-mpg Vs body styles')
```

```python
# 2. Is there a correlation between engine size and horsepower? How does engine size and horsepower vary across different makes and models?
# Solution:
sns.scatterplot(data=dfs,x='engine-size',y='horepower',hue='make')
```

```python
# 3. How does the length, width, and height of the car vary between different body styles and makes?
# Which make and body style has the largest and smallest dimensions?
# Solution:
dim=['length','width','height']
dt_style=df.groupby('body-style')[dim].mean()
dt_make= df.groupby('make')[dim].mean()

dt_style.plot(kind='bar',rot=0)
plt.title('Body styles Vs mean Dimension')
plt.xlabel('Body styles')
plt.ylabel('mean_dimension')

dt_make.plot(kind='bar',rot=90)
plt.title('Makes Vs Mean dimension')
plt.xlabel('make')
plt.ylabel('mean_dimension')
plt.show()
```

```python
#5. Is there a correlation between curb weight and fuel economy (city-mpg and highway-mpg)?
#How does curb weight vary across different body styles and makes?
sns.scatterplot(data=df,y='city-mpg',x='curb-weight')
plt.title('Curb weight Vs City-mpg')
plt.show()
sns.scatterplot(data=df,x='curb-weight',y='highway-mpg')
plt.title('Curb weight Vs Highway-mpg')
plt.show()

corr_city=df['curb-weight'].corr(df['city-mpg'])
corr_highway=df['curb-weight'].corr(df['highway-mpg'])
print('corr_city=',corr_city)
print('corr_highway=',corr_highway)
```

```python
#How does curb weight vary across different body styles and makes?
sns.barplot(data=df,y='curb-weight',x='body-style')
plt.title('Curb weight Vs Body-style')
plt.show()
sns.barplot(data=df,y='curb-weight',x='make')
plt.title('Curb weight Vs Make')
plt.xlabel('make')
plt.xticks(rotation=90)
plt.show()
```
